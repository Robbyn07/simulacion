## Utilizar Matplot y Plotly para realizar las graficas.

```python
##Importar Librerias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io

##Leer el dataset
datos = pd.read_csv('fifa_datos.csv')
datos.head(10)
```

### 1. Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE

* Debajo 125 Lbs.
* 125-150
* 150-175
* 175 o superior

```python
##Limpieza de datos
datos['Weight'] = datos['Weight'].str.replace('lbs', '')
datos['Weight'] = pd.to_numeric(datos['Weight'])

##Toma de datos
debajo = datos.loc[datos['Weight'] <= 125].count()[0]
midBajo = datos.loc[(datos['Weight'] > 125) & (datos['Weight'] <= 150)].count()[0]
midSuper = datos.loc[(datos['Weight'] > 150) & (datos['Weight'] <= 175)].count()[0]
superior = datos.loc[datos['Weight'] > 175].count()[0]

##Grafico
plt.figure()

etiquetas = ['Debajo 125 lbs', '125 - 150 lbs', '150 - 175 lbs', 'Superior 175 lbs']

plt.pie([debajo, midBajo, midSuper, superior], labels=etiquetas, autopct='%.2f %%')
plt.title('Pesos')
plt.show()
```

### 2. Generar un grafico de barras (histograma) de acuerdo a su habilidad (Overall) en base a los siguientes segmentos contando el número de jugadores

* 40
* 50
* 60
* 70
* 80
* 90

```python
over40 = datos.loc[(datos['Overall'] >= 40) & (datos['Overall'] < 50)].count()[0]
over50 = datos.loc[(datos['Overall'] >= 50) & (datos['Overall'] < 60)].count()[0]
over60 = datos.loc[(datos['Overall'] >= 60) & (datos['Overall'] < 70)].count()[0]
over70 = datos.loc[(datos['Overall'] >= 70) & (datos['Overall'] < 80)].count()[0]
over80 = datos.loc[(datos['Overall'] >= 80) & (datos['Overall'] < 90)].count()[0]
over90 = datos.loc[(datos['Overall'] >= 90) & (datos['Overall'] < 100)].count()[0]

plt.figure()

etiquetas = [40, 50, 60, 70, 80, 90]
valores = [over40, over50, over60, over70, over80, over90]

plt.bar(etiquetas, valores)
plt.title('Habilidad')
plt.show()
```
